## Unnecessary Data

Unnecessary data could be irrelevant to your analysis or a duplice column. Check out the dataset to see if there is any unnecessary data.

```python
# Look out for unnecessary data!

# eliminate Age column and rows 1 and 4.

df.drop([1,4]).drop(columns=['Age'])
# eliminate rows with missing data and age column
df.dropna().drop(columns=['Age'])
# drop age column with axis parameter. axis =1 : looks at columns
df.drop(['Age'],axis=1)
```
